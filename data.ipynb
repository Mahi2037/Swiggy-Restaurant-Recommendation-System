```python
import pandas as pd
df = pd.read_csv(r"L:\Guvi\Project 4\cleaned_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Clean Rating
df['rating'] = df['rating'].replace('--', pd.NA)

# Drop rows where rating is missing
df = df.dropna(subset=['rating'])

# Now convert rating to float
df['rating'] = df['rating'].astype(float)

# Clean Rating Count
df['rating_count'] = df['rating_count'].replace('Too Few Ratings', pd.NA)

# Drop rows where rating_count is missing
df = df.dropna(subset=['rating_count'])

# Change K values
# Change K values
def convert_rating_count(x):
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    else:
        return float(x.replace('+ ratings', '').strip())

df['rating_count'] = df['rating_count'].apply(convert_rating_count)

# Convert rating_count to integer
df['rating_count'] = df['rating_count'].astype(int)


# Clean Cost
df['cost'] = df['cost'].str.replace('₹', '', regex=False)
df['cost'] = df['cost'].str.replace(',', '', regex=False)
df['cost'] = df['cost'].astype(float)

# Drop missing values in important columns
df = df.dropna(subset=['name', 'cost', 'cuisine', 'address'])

df.to_csv("cleaned_data.csv", index=False)

print("Data cleaning completed and saved")
```
